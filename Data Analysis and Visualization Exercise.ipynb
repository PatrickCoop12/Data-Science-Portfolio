# Data Analysis and Visualization Exercise

## In this exercise, a small dataset containing stock prices to run some lite analysis and visualizations

```python
# Importing necessary modules/libraries
import pandas as pd
import matplotlib.pyplot as plt
# Creating df dataframe using read_pickle funtion in pandas
df = pd.read_pickle("yahoo_price.pkl")
# Displaying dataframe
df
```

```python
# Showing summary stats using describe function
df.describe()
```

For each company in the data, matplotlib and for loops were used to plot histograms of the price.

```python
# Solve here

# Creation of a position variable to rotate the assignment of histogram plots in our divided graph
position = 1

# Using matplotlib's figure(figsize) to create the figure/frame/size parameters of our graph outside of for loop to avoid redundancies
plt.figure(figsize=(15,15))

# For loop iterating through columns [i] contained in our dataframe
for i in df:
# Subplot function defines our 2x2 grid, position variable used to specify position of graph in our grid
    plt.subplot(2,2,position);
# Putting column values into a histogram, and using column name to generate title. Y axis label is consistent for all graphs 
    plt.hist(df[i])
    plt.title(i + ' Closing Price Histogram')
    plt.ylabel('Closing Price')
    plt.xlabel('Number of days')
# Position increased by 1 after plot is created to specify new grid position for next graph created by for loop
    position = position + 1
```

```python
# The above solution could also be used to create a function if the task was to be repeated. 
# This could be modified to take Fewer inputs if standard graph size, and standard number of rows/columns was desired.

def stock_price_hist(df,subplot_rows,subplot_columns,figxsize,figysize):
    
    position = 1
    plt.figure(figsize=(figxsize,figysize))
    for i in df:
        plt.subplot(subplot_rows,subplot_columns,position);
        plt.hist(df[i])
        plt.title(i + ' Closing Price Histogram')
        plt.ylabel('Closing Price')
        plt.xlabel('Number of days')
        position = position + 1
    return plt.show()
    
stock_price_hist(df,2,2,15,15) 
```
