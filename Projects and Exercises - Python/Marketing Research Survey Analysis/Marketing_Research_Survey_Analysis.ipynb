#### In this project I fielded a small marketing research survey of 78 respondents around generative ai applications, and visualized a perceptual map of trust of and willingness to accept the technology.

```python
index = ['Google Bard','Bing Chat', 'ChatGPT']
percep = {'Trust': [df['trust_bard'].mean(), df['trust_bing_chat'].mean(), df['trust_chatgpt'].mean()],
       'Willingness to Accept': [df['willing_to_acccept_bard'].mean(),df['willing_to_acccept_bing_chat'].mean(), df['willing_to_acccept_chatgpt'].mean()]}

perceptual_map = pd.DataFrame(percep,index = index)
perceptual_map
```

```python
perceptual_map.plot.scatter(x= 'Trust', y= 'Willingness to Accept', s=600)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(perceptual_map['Trust'], perceptual_map['Willingness to Accept'])
for idx, row in perceptual_map.iterrows(): 
    plt.text(row['Trust'], row['Willingness to Accept'], idx,horizontalalignment='center',
     verticalalignment='bottom')
```

```python
plt.scatter(perceptual_map['Trust'], perceptual_map['Willingness to Accept'], s=200)
plt.xlabel('Trust')
plt.ylabel('Willingness to Accept')
plt.axis([0, 4, 0, 4])
#plt.vlines(0,ymin=0, ymax=4)
#plt.hlines(0,xmin=0, xmax=4)
plt.text(perceptual_map['Trust'][0], perceptual_map['Willingness to Accept'][0],'Bard' ,horizontalalignment='left',
     verticalalignment='bottom')
plt.text(perceptual_map['Trust'][1], perceptual_map['Willingness to Accept'][1],'Bing Chat' ,horizontalalignment='left',
     verticalalignment='top')
plt.text(perceptual_map['Trust'][2], perceptual_map['Willingness to Accept'][2],'ChatGPT' ,horizontalalignment='right',
     verticalalignment='top')
```

```python
fig, ax= plt.subplots()
plt.scatter(perceptual_map['Trust'], perceptual_map['Willingness to Accept'], c=['orange', 'blue', 'green'])
plt.xlabel('Trust', fontweight='bold')
plt.ylabel('Willingness to Accept', fontweight='bold')
plt.vlines(2.9,ymin=2.4, ymax=4, colors=['peru'])
#plt.hlines(0,xmin=0, xmax=4)
plt.text(perceptual_map['Trust'][0], perceptual_map['Willingness to Accept'][0],'Bard' ,horizontalalignment='left',
     verticalalignment='top')
plt.text(perceptual_map['Trust'][1], perceptual_map['Willingness to Accept'][1],'Bing Chat' ,horizontalalignment='left',
     verticalalignment='top')
plt.text(perceptual_map['Trust'][2], perceptual_map['Willingness to Accept'][2],'ChatGPT' ,horizontalalignment='right',
     verticalalignment='top')

for axis in ['top','bottom','left', 'right']:
    ax.spines[axis].set_color('peru')
```
