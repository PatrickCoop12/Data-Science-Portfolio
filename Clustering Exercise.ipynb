# Clustering Exercise

## In this exercise a dataset containing salary and occupation data was uploaded then machine learning was used to cluster individuals in the dataset based on years of experience and salaries. The exercise focused on clustering more so than thorough analysis. In the exploratory stage, a scatter plot was used to eye-ball for correlation and direction. In a real-world example, regression analysis would have been conducted to support impressions. 

```python
import pandas as pd 
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Salary Data.csv')
df.dropna(inplace=True)
df
```

```python
plt.scatter(df['Years of Experience'],df['Salary'])
```

```python
df[260:270]
```

```python
km = KMeans(n_clusters=5)
y = km.fit_predict(df[['Years of Experience','Salary']])
y
```

```python
df['cluster'] = y
df.head()
```

```python
# Displaying clusters on graph
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1['Years of Experience'],df1['Salary'], color = 'green')
plt.scatter(df2['Years of Experience'],df2['Salary'], color = 'red')
plt.scatter(df3['Years of Experience'],df3['Salary'], color = 'black')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
```
